### Executive Summary:
In this exercise, two search algorithms are examined, binary search and linear or simple search. Binary search works by taking the middle value and eliminates half the values every time if the search value is not found. On the other hand linear search works by searching one number at a time. The following exercise will show how the two differ by comparing execution times of the algorithms, and explaining in detail how the algorithms work. Further explanation on how search algorithms relate to what is called Big O notation is dicussed and expanded upon. 

5 random data sets were created with different sizes (512, 1024, 2048, 4096, 8192) using a random number generator. Search algorithms were run and execution times were noted followed by a comparison between execution times. Results showed that binary search times were considerably lower than linear search times. It is recommended that binary search algorithms should be used for data engineering applications as higher response times can affect client satisfaction considerably. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Five Arrays of Random Numbers of Different Sizes (512, 1024, 2048, 4096, 8192) Using a Random Number Generator:

```python
#generate random values
from numpy.random import seed
from numpy.random import uniform

#5 arrays of random numbers of different sizes with common seed
seed(156)
array_512 = uniform(1, 100, 512)
seed(156)
array_1024 = uniform(1, 100, 1024)
seed(156)
array_2048 = uniform(1, 100, 2048)
seed(156)
array_4096 = uniform(1, 100, 4096)
seed(156)
array_8192 = uniform(1, 100, 8192)
```

### Sorted Random Arrays with Execution Times:

```python
from timeit import default_timer as timer

#Sort arrays with sort times
start = timer()
array_512 = np.sort(array_512)
end = timer()
sort_execution_time_512 = (end - start)*1000
print('Sorted Execution Time(ms), Size 512:', sort_execution_time_512)

start = timer()
array_1024 = np.sort(array_1024)
end = timer()
sort_execution_time_1024 = (end - start)*1000
print('Sorted Execution Time(ms), Size 1024:', sort_execution_time_1024)

start = timer()
array_2048 = np.sort(array_2048)
end = timer()
sort_execution_time_2048 = (end - start)*1000
print('Sorted Execution Time(ms), Size 2048:', sort_execution_time_2048)

start = timer()
array_4096 = np.sort(array_4096)
end = timer()
sort_execution_time_4096 = (end - start)*1000
print('Sorted Execution Time(ms), Size 4096:', sort_execution_time_4096)

start = timer()
array_8192 = np.sort(array_8192)
end = timer()
sort_execution_time_8192 = (end - start)*1000
print('Sorted Execution Time(ms), Size 8192:', sort_execution_time_8192)
```

### Define Function to Run Binary Search:
 - Big O Run time as **O(log *n*)**, also known as **log time**. 
 - Binary search will will "guess" the middle number and eliminate half the numbers every time based on the "guesses" being either to high or too low*.
 - For example, if there are 100 items, and half the numbers are eliminated every time with binary search, it will take seven steps to find the position of the value (as opposed to worst case being a 100 steps with linear search time). 100 items roughly takes 7ms to run*.  
 - Run time in Big O Notation for Binary Search is log base 2 as log n elements are checked in the worst case*.  
 - Binary search only works if the list is in sorted order.

```python
#Binary Search Base Code
def binary_search(list, item):
  # low and high keep track of which part of the list you'll search in.
    low = 0
    high = len(list) - 1

  # While you haven't narrowed it down to one element ...
    while low <= high:
    # ... check the middle element
        mid = (low + high) // 2
        guess = list[mid]
    # Found the item.
        if guess == item:
            return mid
    # The guess was too high.
        if guess > item:
            high = mid - 1
    # The guess was too low.
        else:
            low = mid + 1

  # Item doesn't exist
    return print('Item does not exist.')
```

### Binary Search Performed on Max Value of Sorted Random Arrays with Execution Times Noted: 

```python
#Binary search with execution time noted in milliseconds

#Size 512
start = timer() #start timer
b_search_array_512 = binary_search(array_512, max(array_512))
end = timer() #end timer
b_execution_time_512 = (end - start)*1000 #convert to milliseconds and store in variable
print('Value Location, Size 512:', b_search_array_512) 
print('Execution time(ms):', b_execution_time_512, '\n')

#Size 1024
start = timer()
b_search_array_1024 = binary_search(array_1024, max(array_1024))
end = timer()
b_execution_time_1024 = (end - start)*1000
print('Value Location, Size 1024:', b_search_array_1024) 
print('Execution time(ms):', b_execution_time_1024, '\n')

#Size 2048
start = timer()
b_search_array_2048 = binary_search(array_2048, max(array_2048))
end = timer()
b_execution_time_2048 = (end - start)*1000
print('Value Location, Size 2048:', b_search_array_2048) 
print('Execution time(ms):', b_execution_time_2048, '\n')

#Size 4096
start = timer()
b_search_array_4096 = binary_search(array_4096, max(array_4096))
end = timer()
b_execution_time_4096 = (end - start)*1000
print('Value Location, Size 4096:', b_search_array_4096) 
print('Execution time(ms):', b_execution_time_4096, '\n')

#Size 8192
start = timer()
b_search_array_8192 = binary_search(array_8192, max(array_8192))
end = timer()
b_execution_time_8192 = (end - start)*1000
print('Value Location, Size 8192:', b_search_array_8192) 
print('Execution time(ms):', b_execution_time_8192, '\n')
```

### Define Function to Run Linear Search or Simple Search:
 - Big O Run time as **O(*n*)**, also known as **linear time**. 
 - Linear search "guesses" one number at a time and eliminates one number at a time.
 - For example, if there are 100 items, if items are a sorted list and the max number is the right number, it will take 100 steps to get to that value. This will roughly take 100ms to run. It is also important to note that Big O does not tell you the speed in seconds. It tells how fast the operation grows*.     
 - Run time in Big O Notation for Linear Search is *n* as each element is checked.  

```python
# Linear Search Base Code
def linear_search(list, item):
    for i in range(len(list)):
        if (list[i] == item):
            return i;    
    return print('Item does not exist');
```

### Linear Search Performed on Max Value of Sorted Random Arrays with Execution Times Noted:

```python
#Linear search with execution time in milliseconds

#Size 512
start = timer() #start timer
l_search_array_512 = linear_search(array_512, max(array_512))
end = timer() #end timer
l_execution_time_512 = (end - start)*1000 #convert to milliseconds and store in variable
print('Value Location, Size 512:', l_search_array_512) 
print('Execution time(ms):', l_execution_time_512, '\n')

#Size 1024
start = timer()
l_search_array_1024 = linear_search(array_1024, max(array_1024))
end = timer()
l_execution_time_1024 = (end - start)*1000
print('Value Location, Size 1024:', l_search_array_1024) 
print('Execution time(ms):', l_execution_time_1024, '\n')

#Size 2048
start = timer()
l_search_array_2048 = linear_search(array_2048, max(array_2048))
end = timer()
l_execution_time_2048 = (end - start)*1000
print('Value Location, Size 2048:', l_search_array_2048) 
print('Execution time(ms):', l_execution_time_2048, '\n')

#Size 4096
start = timer()
l_search_array_4096 = linear_search(array_4096, max(array_4096))
end = timer()
l_execution_time_4096 = (end - start)*1000
print('Value Location, Size 4096:', l_search_array_4096) 
print('Execution time(ms):', l_execution_time_4096, '\n')

#Size 8192
start = timer()
l_search_array_8192 = linear_search(array_8192, max(array_8192))
end = timer()
l_execution_time_8192 = (end - start)*1000
print('Value Location, Size 8192:', l_search_array_8192) 
print('Execution time(ms):', l_execution_time_8192, '\n')
```

### Table of Execution Times by Sort Time, Linear Search Time, Binary Search Time, with Linear Search Time and Sort Time, and with Binary Search Time and Sort Time by Milliseconds:
 - Linear search times are much slower than binary search times.
 - With larger datasets, the differences would be significant.
 - Binary search runs faster as it takes 9 steps to check 512 numbers, 10 steps for 1024, 11 for 2048, 12 for 4096, and 13 for 8192 in worst case scenarios.
 - On the other hand, linear search takes 512 steps for 512 numbers, 1024 steps for 1024, and so on in worst case scenarios. Therefore, it is much slower.
 - This is evident when looking at the search times added with sort times in the table.

```python
times_dataframe = pd.DataFrame(np.array([[len(array_512), sort_execution_time_512, l_execution_time_512, 
                                          b_execution_time_512, l_execution_time_512 + sort_execution_time_512, 
                                          b_execution_time_512 + sort_execution_time_512],
                                        [len(array_1024), sort_execution_time_1024, l_execution_time_1024, 
                                          b_execution_time_1024, l_execution_time_1024 + sort_execution_time_1024, 
                                          b_execution_time_1024 + sort_execution_time_1024],
                                        [len(array_2048), sort_execution_time_2048, l_execution_time_2048, 
                                          b_execution_time_2048, l_execution_time_2048 + sort_execution_time_2048, 
                                          b_execution_time_2048 + sort_execution_time_2048],
                                        [len(array_4096), sort_execution_time_4096, l_execution_time_4096, 
                                          b_execution_time_4096, l_execution_time_4096 + sort_execution_time_4096, 
                                          b_execution_time_4096 + sort_execution_time_4096],
                                        [len(array_8192), sort_execution_time_8192, l_execution_time_8192, 
                                          b_execution_time_8192, l_execution_time_8192 + sort_execution_time_8192, 
                                          b_execution_time_8192 + sort_execution_time_8192]]),
                              columns=['Length_of_Random_Array', 'Sort_Time', 'Linear_Search_Time', 'Binary_Search_Time',  
                                      'Linear_Search_Plus_Sort_Time', 'Binary_Search_Plus_Sort_Time'])
times_dataframe
```

### Plot of Execution Times by Size of Dataset:
 - As discusses in the above, the difference in linear search times versus binary search times in the plot are significant.
 - It is recommended that binary search should be used rather than linear search times, as the significant differences grow with larger dataset sizes.

```python
execution_times_pivot = pd.pivot_table(times_dataframe, index=times_dataframe['Length_of_Random_Array'])
execution_times_plot = execution_times_pivot.loc[:,['Sort_Time','Linear_Search_Time', 'Binary_Search_Time', 
         'Linear_Search_Plus_Sort_Time', 'Binary_Search_Plus_Sort_Time']].plot.bar(stacked=False, figsize=(10,7))
plt.ylabel('Execution Time in Milliseconds')
plt.xlabel('Size of Data Set')
plt.title('Execution Time Against Problem Size by Each Algorithm')
plt.show()
```

### Conclusion:
In this exercise, two search algorithms were examined, binary search and linear or simple search. Binary search works by taking the middle number and eliminates half the numbers every time if the search value is not found based on the "guesses" being either to high or too low. On the other hand, linear search works by searching one number at a time, and will eliminate one number at a time if there is not a match. Binary search cuts down the steps considerably to find the matching value, as a data set with 512 numbers at its worst case scenario will take 9 steps. Linear search at its worst case scenario will take 512 steps with the same dataset. However, binary search requires that the list be in sorted order.  

Binary search and linear search are expressed in *Big O* notation, which tells you how fast an algorithm is. Binary search in Big O notation is written as **O(log *n*)**, also known as **log time**. Additionally, the log in Big O notation always means log base 2. Therefore, 8,000,000,000 items would take around 33 guesses at its worst. On the other hand, linear search in Big O notation is written as **O(*n*)**, which is also known as **linear time**. At its worst linear search would take 8,000,000,000 guesses to find the right value. Big O notation expresses time in the worst case scenario. That is, if a linear search is run and the item searched is found in the first record, it does not mean that it took O(1) time. Therefore, simple search is expressed in **O(*n*)** time. Additionally, Big O does not tell you the speed in seconds; it instead lets you compare the number of operations telling you how fast an algorithm grows*.

The results in the exercise showed the effectiveness of binary search over linear search. The significant difference between binary and linear search times is evidenced with the biggest dataset of 8192 random values. As it only took a maximum of 13 steps to find the right number, as opposed to 8192 steps, this made a a huge impact on execution time. Therefore, it is evident that binary search algorithms should be used in data engineering applications. Just like in the Amazon example described in the above, a slight increase in response time can indeed impact client satisfaction.
